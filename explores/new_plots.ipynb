# box plot

```python
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd
%matplotlib inline

plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["font.sans-serif"] = "SimHei"
```

```python
np.random.RandomState(42)
data_box = np.random.randint(1,100,(100,4))
data_box[1,1] = 200
data_box[3,3] = -100
label_box = ['A', 'B', 'C', 'D']
fig, ax = plt.subplots(1,1, figsize=(3,3))
ax.boxplot(data_box,label_box)
plt.show()
```

## dataframe supported

```python
df_box = pd.DataFrame(data=data_box, columns=label_box)
fig, ax = plt.subplots(1,1, figsize=(3,3))
ax.boxplot(df_box)
plt.show()
```
