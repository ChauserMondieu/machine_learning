# ARMA Simulation
1. <font color=purple><b>arma_generate_sample</b></font>
    > ar  <br>
    > ma  <br>
    > nsample  <br>
    > scale  <br>

```python
from statsmodels.tsa import api as smtsa
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline

# n stands for length of time series
n = 100
# ma defines \theta series
ma = np.r_[1, 0]
# ar defines \phi series
ar1 = np.r_[1, -0.6]
np.random.seed(42)
arma1 = smtsa.arma_generate_sample(ar=ar1, ma=ma, nsample=n,scale=1)

ar2 = np.r_[1, -1]
np.random.seed(43)
arma2 = smtsa.arma_generate_sample(ar=ar2, ma=ma, nsample=n,scale=1)

ar3 = np.r_[1, 1.8]
np.random.seed(44)
arma3 = smtsa.arma_generate_sample(ar=ar3, ma=ma, nsample=n,scale=1)

ar4 = np.r_[1, -1, -0.3]
np.random.seed(45)
arma4 = smtsa.arma_generate_sample(ar=ar4, ma=ma, nsample=n,scale=1)

fig, ax = plt.subplots(2,2, figsize=(12,3), dpi=150,
                       facecolor='whitesmoke', 
                       edgecolor='gray')
fig.subplots_adjust(left=0,right=1,top=1,bottom=0,
                    wspace=0.25,hspace=0.4)

ax[0,0].plot(arma1, 'b-', label="stable")
ax[0,0].legend(loc="best")
ax[0,0].set_xlabel(xlabel = "(a)", fontsize=12)
ax[0,0].set_xticks(ticks=np.round(np.arange(0,101,5),0),fontsize=14)
ax[0,0].set_yticks(ticks=np.round(np.linspace(min(arma1),max(arma1),10),0),fontsize=14)
ax[0,0].grid(alpha=0.4)

ax[0,1].plot(arma2, 'b-', label="random walk")
ax[0,1].legend(loc="best")
ax[0,1].set_xlabel(xlabel = "(b)", fontsize=12)
ax[0,1].set_xticks(ticks=np.round(np.arange(0,101,5),0),fontsize=14)
ax[0,1].set_yticks(ticks=np.round(np.linspace(min(arma2),max(arma2),10),0),fontsize=14)
ax[0,1].grid(alpha=0.4)

ax[1,0].plot(arma3, 'b-', label="unstable")
ax[1,0].legend(loc="best")
ax[1,0].set_xlabel(xlabel = "(c)", fontsize=12)
ax[1,0].set_xticks(ticks=np.round(np.arange(0,101,5),0),fontsize=14)
ax[1,0].set_yticks(ticks=np.round(np.linspace(min(arma3),max(arma3),10),0),fontsize=14)
ax[1,0].grid(alpha=0.4)

ax[1,1].plot(arma4, 'b-', label="unstable")
ax[1,1].legend(loc="best")
ax[1,1].set_xlabel(xlabel = "(d)", fontsize=12)
ax[1,1].set_xticks(ticks=np.round(np.arange(0,101,5),0),fontsize=14)
ax[1,1].set_yticks(ticks=np.round(np.linspace(min(arma4),max(arma4),10),0),fontsize=14)
ax[1,1].grid(alpha=0.4)
plt.show()
```
